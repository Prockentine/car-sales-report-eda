```python
#importing libraries
import pandas as pd
import zipfile
import kaggle
```

```python
#download dataset from Kaggle using Kaggle API
import kagglehub

# Download latest version
path = kagglehub.dataset_download("missionjee/car-sales-report")

print("Path to dataset files:", path)
```

```python
#extract the file from the downloaded zip file
zipfile_name = 'car-sales-report.zip'
with zipfile.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
#read the cvs file as a pandas dataframe
cars = pd.read_csv("Car Sales.xlsx - car_data.csv")
```

```python
#exploring the data
cars.info()
```

```python
#setting Car_id as the index
cars.set_index("Car_id")
```

```python
#exploring how many makes we have data for
cars.Company.value_counts()
```

```python
#exploring the annual income range of customers in our data set
min_income = cars["Annual Income"].min()
max_income = cars["Annual Income"].max()

print(f"Minimum Annual Income: {min_income}")
print(f"Maxiumum Annual Income: {max_income}")
```

```python
#creating a dictionary and renaming some columns for clarity and consistency
new_cols_dict ={
    'Car_id':'Car ID',
    'Date':'Sales Date',
    'Customer Name':'Customer Name',
    'Gender':'Gender',
    'Annual Income':'Annual Income',
    'Dealer_Name':'Dealership',
    'Company':'Make',
    'Model':'Model',
    'Engine':'Engine',
    'Transmission':'Transmission',
    'Color':'Color',
    'Price ($)':'Sales Price',
    'Dealer_No':'Dealer Number',
    'Body Style':'Body Style',
    'Phone':'Customer Phone',
    'Dealer_Region':'Dealer Region',
}

#renaming the columns to my new column names
cars.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#checking to ensure our renaming worked
cars.head()
```

```python
#converting dtypes for all objects to strings
cars = cars.astype({col: "str" for col in cars.select_dtypes(include="object").columns})
```

```python
#checking to ensure our dtypes are updated
print(cars.dtypes)
```

```python
#writing the dataframe to an xlsx file for Tableau
cars.to_excel('car_sales_data.xlsx', sheet_name='Sales Data')
```
